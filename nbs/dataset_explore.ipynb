```python
import pandas as pd
import json
from collections import Counter
```

```python
with open("../data/kan/kan_train.json") as fob:
    data = [json.loads(x) for x in fob.read().strip().split("\n")]
df = pd.DataFrame(data)
df
```

```python
counters = Counter(''.join(list(df['native word'])))
```

```python
for c in counters.most_common():
    print(c, ord(c[0]))
```

```python
sorted([ord(c[0]) for c in counters])
```
